# Data and Source

## Dependent Variables

| **Dependent variable** | **Why** | **Level available** | **Source** |
|---|---|---|---|
| **Mobility rate** | Best direct measure of lock-in because homeowners become less likely to move when giving up a low-rate mortgage is costly. Census ACS mobility tables are based on whether people lived in the same residence one year ago. | National, state, county, place | [U.S. Census ACS mobility table B07001](https://data.census.gov/table/ACSDT1Y2024.B07001?q=B07001) |
| **Existing-home sales** | Strong market-level outcome for lock-in. If owners are reluctant to give up favorable financing, transaction volume in existing homes should fall. | National, Census region | [FRED: Existing Home Sales (EXHOSLUSM495S)](https://fred.stlouisfed.org/series/EXHOSLUSM495S) |
| **New listing count** | Captures the “not even putting the house on the market” margin of lock-in. When owners do not want to give up favorable mortgage terms, listing activity may fall. | National, state, metro, county, ZIP code | [Realtor.com Research Data Library](https://www.realtor.com/research/data/) |

**Notes**

- **Mobility rate:** ACS mobility tables are available through Census products covering multiple geographies.  
- **Existing-home sales:** The FRED/NAR series is available at the national and Census-region levels.  
- **New listing count:** Realtor.com provides listing and inventory data across several geographic levels.

# Outline

## Study focus

- Examine whether the **housing lock-in effect** is associated with lower housing market activity
- Study whether this effect is stronger under different:
  - **unemployment conditions**
  - **industry structures**

A study of the lock-in effect helps businesses understand whether weak housing activity reflects a true lack of demand or a financial friction that changes customer behavior, market turnover, and regional opportunity.

## Data to be used

- **National time-series data** as the main dataset
- Monthly or quarterly measures of:
  - **existing-home sales** or **new listing count**
  - **mortgage lock-in measure**
  - **unemployment rate**
  - **industry-related measure**
  - **house price growth**
  - **supply-related measure** such as inventory or permits

## Main variables

- **Outcome**
  - mobility rate
  - existing-home sales
  - new listing count

- **Main predictor**
  - mortgage lock-in measure

- **Main effect modifiers**
  - unemployment rate
  - industry structure, represented by grouped states as a simplified proxy

- **Key controls**
  - house price growth
  - supply conditions

## Models to be used

- **SARIMAX** as the baseline model
- **XGBoost** as the nonlinear machine learning model
- **LSTM** as the sequence-based deep learning model

## Model setup

- Use the housing outcome as the target variable
- Include:
  - lock-in measure
  - unemployment rate
  - industry measure
  - lock-in × unemployment
  - lock-in × industry
  - selected control variables
- Add lagged terms and seasonal structure where needed

## Evaluation plan

- Use **SARIMAX as the benchmark**
- Compare models using:
  - RMSE
  - MAE
  - MAPE if appropriate
- Use **rolling time-series validation**
- Compare both:
  - predictive performance
  - interpretability of the lock-in effect

## Expected contribution

- Provide an interpretable baseline estimate of the lock-in effect
- Test whether **unemployment** and **industry structure** strengthen or weaken this effect
- Compare traditional time-series modeling with machine learning approaches

# Models and Details

### 1) Research objective and modeling target

The goal of this study is to examine whether **industry structure** and the **unemployment rate** are associated with the strength of the **housing lock-in effect**, where lock-in refers to the reduced willingness of homeowners to move, sell, or list their homes when current mortgage rates are much higher than the rates embedded in existing mortgages.

The main modeling target is a national time-series housing outcome, such as:

- **Existing-home sales**
- **New listing count**
- **Mobility rate** (if the time frequency is sufficient)

Let the outcome be denoted by $Y_t$, and let the main lock-in measure be denoted by $L_t$.

---

### 2) Covariates

A compact set of covariates will be included to account for macroeconomic and housing-market conditions that may affect the outcome independently of lock-in. These covariates include:

- **House price pressure**, denoted by $H_t$
- **Supply conditions** such as inventory or building permits, denoted by $S_t$
- **Unemployment rate**, denoted by $U_t$
- **Industry structure**, denoted by $I_t$

The main focus is on whether $U_t$ and $I_t$ modify the effect of $L_t$.

---

### 3) Models that are going to use and their assumptions

#### (1) SARIMAX baseline model

SARIMAX will be used as the primary baseline model because it can jointly account for:

- serial dependence,
- seasonality,
- exogenous predictors.

General form:

$$
Y_t = \beta_0 + \sum_{j=1}^{p} \phi_j Y_{t-j} + \sum_{k=1}^{K} \beta_k X_{k,t} + \sum_{m=1}^{q} \theta_m \varepsilon_{t-m} + \varepsilon_t
$$

with possible seasonal structure added at lag $s$.

**Main assumptions:**
- the time-series dependence is adequately captured by the ARIMA/SARIMA structure,
- the relationship between predictors and outcome is approximately linear,
- residual autocorrelation is sufficiently removed after model fitting,
- the series is stationary after appropriate transformation or differencing,
- no major omitted time-varying confounders.

---

#### (2) XGBoost

XGBoost will be used as a nonlinear machine-learning extension to capture:

- nonlinear effects,
- threshold behavior,
- higher-order interactions,
- flexible lag structures.

Abstract form:

$$
\hat{Y}_t = f(\mathbf{Z}_t)
$$

where $\mathbf{Z}_t$ is a feature vector containing lagged outcomes, lagged predictors, rolling summaries, and seasonal indicators.

**Main assumptions:**
- the training and testing data come from a reasonably stable data-generating process,
- there is enough data to learn nonlinear relationships,
- the engineered features capture the relevant time dependence,
- no future information leaks into the predictors.

---

#### (3) LSTM

LSTM will be used as a sequence-based deep learning model to learn temporal dependencies directly from ordered input sequences.

Abstract form:

$$
\hat{Y}_t = g(\mathbf{Z}_{t-w+1}, \mathbf{Z}_{t-w+2}, \dots, \mathbf{Z}_t)
$$

where $w$ is the sequence window length.

**Main assumptions:**
- the available time series is long enough for stable sequence learning,
- the sequential structure contains meaningful predictive information,
- preprocessing preserves temporal order,
- no information leakage occurs across train/test periods,
- the model is not severely overfitted.

---

### 4) Variable selection for each model

#### SARIMAX

Variable selection for SARIMAX will be theory-driven and parsimonious. The initial candidate set will include:

- lock-in measure: $L_t$
- unemployment: $U_t$
- industry structure: $I_t$
- house price pressure: $H_t$
- supply conditions: $S_t$
- interaction terms:
  - $L_t \times U_t$
  - $L_t \times I_t$

Lagged versions of the predictors may also be considered, such as $L_{t-1}$, $U_{t-1}$, and $H_{t-1}$, depending on data frequency and theory.

Selection will be based on:
- economic interpretability,
- information criteria such as AIC or BIC,
- residual diagnostics,
- parsimony.

---

#### XGBoost

Variable selection for XGBoost will be feature-based. Candidate features will include:

- lagged outcome values: $Y_{t-1}, Y_{t-2}, \dots$
- lagged lock-in variables: $L_{t-1}, L_{t-2}, \dots$
- lagged unemployment and industry variables
- lagged house price and supply variables
- rolling averages and rolling changes
- seasonal indicators such as month or quarter
- interaction features if needed

The final feature set will be refined using:
- out-of-sample validation,
- feature importance,
- permutation importance,
- SHAP-based screening if needed.

---

#### LSTM

Variable selection for LSTM will be sequence-based rather than coefficient-based. The input at each time step will include:

- $L_t$
- $U_t$
- $I_t$
- $H_t$
- $S_t$
- possibly $Y_t$ history within the sequence window

Feature selection will emphasize:
- low redundancy,
- consistent scaling,
- stable sequence construction,
- predictive contribution in validation performance.

Because LSTM is more data-hungry, the variable set should remain relatively compact.

---

### 5) Model specification

#### SARIMAX baseline specification

A baseline specification can be written as:

$$
Y_t = \beta_0 + \beta_1 L_t + \beta_2 U_t + \beta_3 I_t + \beta_4 H_t + \beta_5 S_t + \beta_6 (L_t \times U_t) + \beta_7 (L_t \times I_t) + n_t
$$

where $n_t$ follows a seasonal ARIMA process.

If lagged predictors are included, the specification can be extended to:

$$
Y_t = \beta_0 + \sum_{\ell=0}^{r_L} \alpha_{\ell} L_{t-\ell}
+ \sum_{\ell=0}^{r_U} \gamma_{\ell} U_{t-\ell}
+ \sum_{\ell=0}^{r_I} \delta_{\ell} I_{t-\ell}
+ \sum_{\ell=0}^{r_H} \eta_{\ell} H_{t-\ell}
+ \sum_{\ell=0}^{r_S} \kappa_{\ell} S_{t-\ell}
+ \lambda_1 (L_t \times U_t)
+ \lambda_2 (L_t \times I_t)
+ n_t
$$

---

#### XGBoost specification

XGBoost will use a tabular feature matrix built from lagged and engineered predictors:

$$
\hat{Y}_t = f\big(Y_{t-1}, \dots, Y_{t-p}, L_t, \dots, L_{t-r}, U_t, \dots, U_{t-r}, I_t, \dots, I_{t-r}, H_t, S_t, \text{seasonal features}\big)
$$

The interaction effects do not need to be manually imposed, since tree-based models can learn them automatically, although selected explicit interactions may also be added.

---

#### LSTM specification

LSTM will use an input sequence of length $w$:

$$
\hat{Y}_t = g\left(
\begin{bmatrix}
L_{t-w+1}, U_{t-w+1}, I_{t-w+1}, H_{t-w+1}, S_{t-w+1} \\
L_{t-w+2}, U_{t-w+2}, I_{t-w+2}, H_{t-w+2}, S_{t-w+2} \\
\vdots \\
L_t, U_t, I_t, H_t, S_t
\end{bmatrix}
\right)
$$

Optionally, lagged values of the outcome may also be included within the sequence.

---

### 6) Evaluation metrics comparing to the baseline model

The SARIMAX model will be treated as the baseline benchmark. All models will be compared using **time-series validation**, such as rolling-origin or expanding-window evaluation.

The main performance metrics will include:

- **Root Mean Squared Error (RMSE)**:
$$
\text{RMSE} = \sqrt{\frac{1}{n}\sum_{t=1}^{n}(Y_t - \hat{Y}_t)^2}
$$

- **Mean Absolute Error (MAE)**:
$$
\text{MAE} = \frac{1}{n}\sum_{t=1}^{n} |Y_t - \hat{Y}_t|
$$

- **Mean Absolute Percentage Error (MAPE)**, if appropriate:
$$
\text{MAPE} = \frac{100}{n}\sum_{t=1}^{n}\left|\frac{Y_t - \hat{Y}_t}{Y_t}\right|
$$

Models will be compared against the SARIMAX baseline in two ways:

1. **Predictive performance**
   - lower RMSE, MAE, or MAPE than SARIMAX

2. **Substantive usefulness**
   - whether the model helps identify how unemployment and industry structure modify the lock-in effect

For interpretation:

- **SARIMAX**: use coefficient estimates, confidence intervals, and interaction effects
- **XGBoost**: use feature importance, permutation importance, SHAP values, and partial dependence plots
- **LSTM**: use predictive accuracy, ablation analysis, and sensitivity checks

The final model assessment should balance:
- predictive improvement over the baseline,
- interpretability,
- and consistency with the substantive housing-lock-in mechanism.
