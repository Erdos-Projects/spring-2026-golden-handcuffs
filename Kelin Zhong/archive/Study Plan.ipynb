## Definition & Measurement

**Lock-In Effect**: The Lock-In Effect (or "Golden Handcuffs") occurs when a homeowner is financially discouraged from selling their current home because their existing mortgage rate is significantly lower than the rate they would receive on a new purchase. This "locks" them in place, even if they have a life reason to move (e.g., job change, growing family).


### What measurement should we choose to measure the lock-in effect?




### Aggregate Measurement

**Rate Gap 10-year ago**:

$$
\Delta r_{10yr,t} = R_{m,t} - R_{10yr,t}
$$

**Rate Gap**:

$$
\Delta r_{t} = R_{m,t} - R_{p,t}
$$

- $R_{m,t}$: 30-year fixed mortage rate
- $R_{p,t}$: Weighted average rate on outstanding mortgages derived from the National Mortgage Database (NMDB) Outstanding Residential Mortgage Statistics (Houston MSA slice if available; otherwise national/state).

Interpretation: higher $\Delta r$ indicate stronger lock-in; you expect fewer new listings and lower turnover controlling for fundamentals.

**Mortgage–Treasury Spread:**: 

$$
S_{t} = R_{m, t} - Y_{10, t}
$$

When $S_{t}$ is above its historical norm, it indicates that mortgage rates are unusually high relative to Treasuries, often due to credit risk, prepayment risk, or investor risk-aversion. Even if the 10-year yield falls, a sticky wide spread can keep $R_{m}$ high, sustaining lock-in.

**Turnover Velocity / Listing Propensity**: 

1) New Listing Rate:

$$
\text{New Listing Rate}_{t} = \frac{\text{New Listings}_{t}}{\text{Housing Stock}_{t}}
$$

- New Listings: New Listing Count = FRED (NEWLISCOU26420).
- Housing Stock: Total housing units in the Houston MSA from ACS (e.g., DP04 table).

2) Inventory Turnover (for robustness):

$$
\text{Turnover}_{t} = \frac{\text{Existing Home Sales}_{t}}{\text{Housing Stock}_{t}}
$$

- Sales: TRERC MSA Housing Report / HAR MLS.

3) New Listing Growth Rate:

$$
\text{New Listing Growth Rate}_{t} = \frac{\Delta \text{New Listing}_{t}}{\text{New Listing}_{t-1}}
$$
