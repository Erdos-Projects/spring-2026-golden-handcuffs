# Research Plan

## Study Focus

This study examines whether the **housing lock-in effect** is associated with lower new listing activity, and whether this effect is strengthened or weakened under different **unemployment conditions** and **industry structures**.

The lock-in effect refers to the reduced willingness of homeowners to list their homes for sale when current mortgage rates are substantially higher than the rates embedded in their existing mortgages. A homeowner who refinanced at 3% faces a large financial penalty from giving up that rate to purchase a new home at 7%, effectively discouraging them from listing even when life circumstances would otherwise motivate a move.

Understanding this mechanism matters beyond academic interest. If weak listing activity reflects a financial friction rather than a true absence of supply, then businesses, policymakers, and investors should interpret low listing volumes differently — as a latent supply problem with a structural cause, not a permanent supply collapse.

## Modeling Structure

The study is organised around **two separate levels of analysis**, each answering a distinct research question with a correspondingly appropriate data structure.

**National level model** examines whether the lock-in effect reduces new listing activity at the aggregate level, and whether this effect is amplified when unemployment is high. This model uses a single national time series at monthly frequency from approximately 2000 to 2024, yielding roughly 288 monthly observations. The unemployment interaction $\tilde{L}_t \times \tilde{U}_t$ is identified from time-series variation in both variables at the national level.

**State level model** examines whether the lock-in effect differs across states with different dominant industry structures, and whether the unemployment interaction persists at the sub-national level. This model uses a pooled panel of 50 states at monthly frequency from approximately 2000 to 2024, yielding roughly 6,000 state-month observations. The model includes state fixed effect dummies, industry group dummies $D_{ki}$, and interaction terms $\tilde{L}_t \times D_{ki}$ directly in the model — one pooled model with dummy variables handling all cross-sectional structure. The industry interaction is identified from cross-sectional variation in $D_{ki}$ across states — states in different industry groups face the same national $L_t$ simultaneously, making the differential response attributable to industry structure.

The two levels are complementary rather than competing. The national model establishes the aggregate lock-in effect and its unemployment modifier with maximum statistical power on a clean single time series. The state model tests whether the effect is heterogeneous across industry structures using the richer variation in the panel. Convergent results across both levels and both models constitute strong evidence. Results that differ across levels are themselves informative about the scale at which the mechanism operates.

Two models are applied at each level:

- **SARIMAX** as the parametric econometric baseline with explicit serial and seasonal structure, producing interpretable coefficient estimates with confidence intervals
- **XGBoost** as the nonlinear machine learning complement, capturing threshold behaviour and flexible interactions through SHAP-based interpretation

# Models and Details

## 1) Research Objective and Modeling Target

The goal is to examine whether **unemployment** and **industry structure** modify the strength of the housing lock-in effect on new listing activity.

At the national level, the modeling target is $Y_t$, the national monthly new listing count. At the state level, the modeling target is $Y_{it}$, the monthly new listing count for state $i$ in month $t$. The main predictor is $L_t$ — the mortgage rate premium — in both cases.

## 2) Covariates

**Both levels:**
- House price pressure $H_t$ / $H_{it}$: FHFA monthly HPI
- Supply conditions $S_t$ / $S_{it}$: building permits from Census BPS
- Unemployment rate $U_t$ / $U_{it}$: BLS LAUS monthly series

**State level only:**
- Industry group dummy $D_{ki}$: fixed label per state, assigned from BLS CES 2000–2005 base period, 3 or 4 groups maximum

## 3) Data Treatment Before Modeling

### 3a) Mean-Centering for SARIMAX Interaction Terms

Variables entering interaction terms in SARIMAX are mean-centered on the training window. Centering is applied **after** the stationarity transformation — that is, to the same form of each variable that actually enters the model.

**For the mortgage rate premium $L_t$:** the national model uses the **first difference** $\Delta L_t = L_t - L_{t-1}$ rather than the level of $L_t$. The reason is stationarity — $L_t$ in levels is non-stationary over the 2016–2025 window (it trended sharply upward from 2021 to 2023), whereas $\Delta L_t$ is stationary. The first difference is then centered on its training-window mean:

$$\widetilde{\Delta L}_t = \Delta L_t - \overline{\Delta L}$$

The centering mean $\overline{\Delta L}$ is computed on the training period only and applied mechanically to the test period, preventing data leakage.

**For the unemployment rate $U_t$:** the level of $U_t$ is stationary (confirmed by ADF and KPSS in the national model notebook), so it enters in levels. It is centered on its training-window mean:

$$\tilde{U}_t = U_t - \bar{U}$$

The full set of centered regressors entering the SARIMAX exog block is therefore:

$$\left[\widetilde{\Delta L}_t,\; \tilde{U}_t,\; \Delta H_t,\; Delta S_t,\; \widetilde{\Delta L}_t \times \tilde{U}_t\right]$$

Note that $H_t$ and $S_t$ also enter in first-differenced form ($\Delta H_t$, $\Delta S_t$) consistent with the stationarity treatment applied to the model, but they are not centered because they do not enter any interaction term.

**Coefficient interpretation with centering:**

- $\beta_1$: effect of a one-unit change in $\widetilde{\Delta L}_t$ on new listings, evaluated at mean unemployment (i.e., when $\tilde{U}_t = 0$)
- $\beta_2$: effect of a one-unit increase in $\tilde{U}_t$ on new listings, evaluated when $\widetilde{\Delta L}_t = 0$ (i.e., at the mean rate of change of the lock-in premium)
- $\beta_5$: how much the lock-in effect changes per unit of $\tilde{U}_t$ above its mean — a negative $\beta_5$ indicates the lock-in effect is stronger when unemployment is elevated

**Why center at all?** Without centering, the main effect coefficients $\beta_1$ and $\beta_2$ are evaluated at zero of the other variable, which is outside the data range and economically meaningless. Centering shifts the evaluation point to the sample mean, making $\beta_1$ and $\beta_2$ directly interpretable as "typical" effects.

All centering constants are computed on the **training window only** and applied mechanically to the test period to prevent data leakage. $D_{ki}$ does not enter interactions in SARIMAX and is kept in original scale.

For **XGBoost**: original scale throughout. Tree splits are rank-invariant so centering adds no benefit and reduces interpretability of SHAP values.

### 3b) Stationarity

All time-series and panel variables are tested for unit roots before modeling:

1. Apply **ADF test** and confirm with **KPSS test** on each variable
2. If a unit root is found, apply first differencing $\Delta Y_t = Y_t - Y_{t-1}$
3. Re-test the differenced series to confirm stationarity
4. The differencing decision is made on the training data and applied mechanically to validation and test splits

For SARIMAX: if $Y_t$ requires differencing ($d \geq 1$), each exogenous regressor is transformed to its stationary form (level or first difference as determined above) before centering. The stationarity transformation is applied to both the outcome and the regressors; centering is a second step applied after the stationarity transformation.

For XGBoost: apply the same differencing decisions used in SARIMAX for consistency across models.

### 3c) Mortgage Rate Premium Construction

$$L_t = r_t^{\text{MORTGAGE30US}} - r_t^{\text{GS10}}$$

where $r_t^{\text{MORTGAGE30US}}$ is the monthly average of weekly FRED MORTGAGE30US observations within each calendar month, and $r_t^{\text{GS10}}$ is the FRED GS10 monthly series ready as-is. No interpolation required. $L_t$ is a national series applied uniformly to all states.

Validate against FHFA NMDB quarterly outstanding rate over the available overlap period. If the correlation falls below 0.85, substitute the weighted average construction using a 7-year exponential decay of MORTGAGE30US.

### 3d) Industry Group Assignment

State industry groups are assigned from BLS CES employment shares averaged over 2000–2005, before the main study period begins. Pre-determination avoids endogeneity between industry composition and listing activity during the study period. Groups are limited to 3 or 4 categories — for example manufacturing-heavy, services-heavy, and mixed — to preserve degrees of freedom and interpretability.

### 3e) Seasonal Adjustment

New listing count exhibits strong monthly seasonality. SARIMAX handles this directly within the SARIMA component at seasonal lag $s = 12$. For XGBoost, seasonal structure is controlled by adding month-of-year dummies or Fourier sine/cosine pairs as explicit features in the feature matrix.

## 4) National Level Models

### National Model 1: SARIMAX

**Purpose:** parametric baseline for the national time series. Produces directly interpretable coefficient estimates with confidence intervals for the lock-in effect and its interaction with unemployment.

**General form:**

$$Y_t = \beta_0 + \sum_{j=1}^{p} \phi_j Y_{t-j} + \sum_{k=1}^{K} \beta_k X_{k,t} + \sum_{m=1}^{q} \theta_m \varepsilon_{t-m} + \varepsilon_t$$

with seasonal structure at lag $s = 12$.

**Main assumptions:**
- serial dependence is adequately captured by the ARIMA/SARIMA structure
- the relationship between predictors and outcome is approximately linear
- residuals are white noise after model fitting
- the series is stationary after transformation or differencing
- no major omitted time-varying confounders

**Additional Note**: Non-stationarity in the outcome is addressed through the integrated SARIMA structure; exogenous regressors are entered in their stationary form (level or first difference as determined by ADF/KPSS), with robustness checks for alternative transformed specifications when needed.

**Baseline specification:**

Stationarity testing on the training set determines the form of each regressor. In the national model notebook, $L_t$ (the lock-in premium) requires first differencing to achieve stationarity, while $U_t$ (unemployment) is stationary in levels. Centering is then applied to the variables that enter the interaction term:

$$\widetilde{\Delta L}_t = \Delta L_t - \overline{\Delta L}, \qquad \tilde{U}_t = U_t - \bar{U}$$

where all means are computed on the training window only. The national SARIMAX exog block is:

$$\left[\widetilde{\Delta L}_t,\; \tilde{U}_t,\; \Delta H_t,\; \Delta S_t,\; \widetilde{\Delta L}_t \times \tilde{U}_t\right]$$

and the model equation is:

$$Y_t = \beta_0 + \beta_1 \widetilde{\Delta L}_t + \beta_2 \tilde{U}_t + \beta_3 \Delta H_t + \beta_4 \Delta S_t + \beta_5 (\widetilde{\Delta L}_t \times \tilde{U}_t) + n_t$$

where $n_t \sim \text{SARIMA}(p, d, q)(P, D, Q)_{12}$.

**Coefficient interpretation:**
- $\beta_1$: effect of a one-unit increase in $\widetilde{\Delta L}_t$ on new listings at mean unemployment
- $\beta_2$: effect of a one-unit increase in $\tilde{U}_t$ on new listings when the lock-in premium is changing at its mean rate
- $\beta_5$: how much the lock-in effect strengthens or weakens per unit of $\tilde{U}_t$ above its mean. A negative $\beta_5$ indicates the lock-in effect is stronger when unemployment is high.

**Model selection procedure:**
1. Test $Y_t$ for stationarity using ADF and KPSS. Apply first differencing if required.
2. Test each regressor ($L_t$, $U_t$, $H_t$, $S_t$) for stationarity on the training set. Apply first differencing to any non-stationary regressor.
3. Select ARIMA order $(p, d, q)$ and seasonal order $(P, D, Q)_{12}$ on $Y_t$ alone using AIC or BIC, without exogenous regressors.
4. Add the full exogenous regressor set in stationary, centered form and re-estimate.
5. Verify residual white noise via Ljung-Box test and inspect ACF/PACF of residuals.
6. Retain $\widetilde{\Delta L}_t \times \tilde{U}_t$ regardless of marginal significance.
7. Keep distributed lag depth at $r \leq 2$ if lagged predictors are added.

**The selected ARIMA order $(p, d, q)(P, D, Q)_{12}$ from the national model is carried forward as the common order for the state-level SARIMAX model.**

### National Model 2: XGBoost

**Purpose:** nonlinear ML complement to SARIMAX. Captures threshold behaviour and flexible interactions without imposing a linear structure. SHAP values provide the primary interpretation of the lock-in effect and its interaction with unemployment.

**Abstract form:**

$$\hat{Y}_t = f(\mathbf{Z}_t)$$

where $\mathbf{Z}_t$ is a feature vector containing the same theoretically motivated variables that enter the SARIMAX exog block, plus a seasonal lag of $Y_t$.

**Feature construction:**

$$\mathbf{Z}_t = \big(Y_{t-12},\; \widetilde{\Delta L}_t,\; \tilde{U}_t,\; \Delta H_t,\; \Delta S_t,\; \widetilde{\Delta L}_t \times \tilde{U}_t\big)$$

The feature set is deliberately restricted to the same six economically motivated variables that enter the SARIMAX exog block. This is not a comparability-of-inputs claim — SARIMAX also uses its internal ARIMA structure, which XGBoost does not have. The restriction serves a different purpose: it keeps the XGBoost interpretation focused on the economic mechanism rather than pure time-series pattern matching, and makes the directional consistency check across the two models meaningful. Both models are asked the same economic question using the same theoretically justified regressors.

$Y_{t-12}$ is added as a seasonal anchor. It gives XGBoost access to the prior-year level of the outcome — the same seasonal memory that SARIMAX captures through its $(P, D, Q)_{12}$ component — without introducing a long lag structure that would consume degrees of freedom on the ~90-observation training set.

All features are constructed in the same transformed form used in the SARIMAX exog block ($\widetilde{\Delta L}_t$, $\tilde{U}_t$, $\Delta H_t$, $\Delta S_t$). Centering carries no benefit for XGBoost since tree splits are rank-invariant, but it is retained so that SHAP values for $\widetilde{\Delta L}_t$ and $\tilde{U}_t$ are directly comparable in sign and direction to the SARIMAX coefficients $\beta_1$ and $\beta_2$.

**Hyperparameter selection:**

| Hyperparameter | Value | Rationale |
|---|---|---|
| `n_estimators` | 200–500 with early stopping | Early stopping determines effective number of trees |
| `max_depth` | 3–4 | Shallow trees reduce overfitting on ~90 training rows |
| `learning_rate` | 0.01–0.05 | Small rate with early stopping is more stable |
| `subsample` | 0.7–0.8 | Row subsampling reduces variance |
| `colsample_bytree` | 0.6–0.8 | Column subsampling reduces feature correlation |
| `min_child_weight` | 5–10 | Prevents splits on very small node groups at small N |
| `reg_alpha` | 0.5 | L1 regularisation; fixed given small N |
| `reg_lambda` | 2.0 | L2 regularisation; fixed given small N |

**Tuning procedure:**
1. Tune `max_depth`, `min_child_weight`, `subsample`, and `colsample_bytree` jointly using time-series cross-validation with an expanding window (3 folds given ~90 training rows — 5 folds would produce initial folds too small for stable estimates).
2. Fix `n_estimators=200` during CV for stability; apply early stopping only in the final fit on the full training set.
3. `reg_alpha` and `reg_lambda` are fixed rather than tuned: with only ~90 training rows, a second-pass regularisation search adds variance without reliable signal.
4. All cross-validation folds must respect temporal order.

**Interpretation outputs:**
- SHAP main effect for $\widetilde{\Delta L}_t$: mean absolute SHAP value and sign across test set
- SHAP dependence plot for $\widetilde{\Delta L}_t$ coloured by $\tilde{U}_t$: visualises the interaction
- SHAP dependence plot for $\tilde{U}_t$ coloured by $\widetilde{\Delta L}_t$: investigates the unemployment conflict
- SHAP bar and beeswarm plots for overall feature importance ranking

**The feature set and hyperparameter ranges from the national model serve as the starting point for the state-level XGBoost model.**

## 5) State Level Models

### State Model 1: SARIMAX

**Purpose:** extends the national SARIMAX to the state-level pooled panel to test whether the lock-in effect differs across industry groups and whether the unemployment interaction persists at the sub-national level. One single pooled SARIMAX model is estimated on all 50 states with state fixed effect dummies, industry group dummies, and interaction terms handling all cross-sectional structure. The common ARIMA order from the national model is imposed directly.

**State-level SARIMAX specification:**

Mean-center the interaction variables using training-window means, after applying the same stationarity transformations established in the national model. Since $L_t$ requires first differencing, the centered form used in the state model is:

$$\widetilde{\Delta L}_t = \Delta L_t - \overline{\Delta L}, \qquad \tilde{U}_{it} = U_{it} - \bar{U}_i$$

where $\overline{\Delta L}$ is the mean of $\Delta L_t$ over the training window and $\bar{U}_i$ is the training-window mean of unemployment for state $i$.

The pooled state-level SARIMAX equation is:

$$Y_{it} = \alpha_i + \beta_1 \widetilde{\Delta L}_t + \beta_2 \tilde{U}_{it} + \beta_3 \Delta H_{it} + \beta_4 \Delta S_{it} + \beta_5 (\widetilde{\Delta L}_t \times \tilde{U}_{it}) + \sum_{k=1}^{m-1} \gamma_k D_{ki} + \sum_{k=1}^{m-1} \lambda_k (\widetilde{\Delta L}_t \times D_{ki}) + n_{it}$$

where:
- $\alpha_i$ are state fixed effect dummies absorbing time-invariant state heterogeneity in baseline listing activity
- $D_{ki}$ are industry group dummies varying across states — this cross-sectional variation is what identifies $\lambda_k$
- $n_{it} \sim \text{SARIMA}(p, d, q)(P, D, Q)_{12}$ using the common order from the national model
- $L_t$ has no $i$ subscript — the mortgage rate premium is national and identical for all states in each period

**Coefficient interpretation:**
- $\beta_1$: effect of a one-unit increase in $\widetilde{\Delta L}_t$ on new listings at mean state unemployment, for the reference industry group
- $\beta_5$: how much the lock-in effect changes per unit of $\tilde{U}_{it}$ above its state mean
- $\gamma_k$: baseline listing activity difference for industry group $k$ relative to the reference group
- $\lambda_k$: how much the lock-in effect differs in industry group $k$ relative to the reference group. A negative $\lambda_k$ for manufacturing-heavy states indicates the lock-in effect is stronger there.

**Model selection procedure:**
1. Impose the common ARIMA order $(p, d, q)(P, D, Q)_{12}$ from the national model directly — no re-selection.
2. Add the full exogenous regressor set including state dummies $\alpha_i$, $\widetilde{\Delta L}_t$, $\tilde{U}_{it}$, $\Delta H_{it}$, $\Delta S_{it}$, $\widetilde{\Delta L}_t \times \tilde{U}_{it}$, $D_{ki}$, and $\widetilde{\Delta L}_t \times D_{ki}$.
3. Verify residual white noise via Ljung-Box test.
4. If the model shows significant residual autocorrelation under the common order, note this as a limitation and report an unconstrained version as a robustness check.
5. Retain all interaction terms $\widetilde{\Delta L}_t \times \tilde{U}_{it}$ and $\widetilde{\Delta L}_t \times D_{ki}$ regardless of marginal significance.

### State Model 2: XGBoost

**Purpose:** extends the national XGBoost to the state-level pooled panel. Tests whether the lock-in effect is nonlinearly modified by industry structure and unemployment. SHAP interaction values between $L_t$ and $D_{ki}$ provide the primary evidence for the industry interaction, complementing the SARIMAX coefficient estimates.

**Data structure:** pooled panel of all 50 states × monthly observations, approximately 6,000 state-month rows.

**Abstract form:**

$$\hat{Y}_{it} = f(\mathbf{Z}_{it})$$

**Feature construction — extends the national feature set:**

$$\mathbf{Z}_{it} = \big(Y_{i,t-12},\; \widetilde{\Delta L}_t,\; \tilde{U}_{it},\; \Delta H_{it},\; \Delta S_{it},\; \widetilde{\Delta L}_t \times \tilde{U}_{it},\; \alpha_i,\; D_{ki},\; \widetilde{\Delta L}_t \times D_{ki}\big)$$

Key additions relative to the national model:
- **State fixed effect dummies** $\alpha_i$ absorb time-invariant state-level heterogeneity in baseline listing activity
- **Industry group dummies** $D_{ki}$ capture baseline differences across industry groups
- **Explicit interaction features** $\widetilde{\Delta L}_t \times D_{ki}$ make the industry interaction directly visible in SHAP output
- $Y_{i,t-12}$ is computed within each state's own time series to avoid cross-state contamination

**Hyperparameter selection:**

Start from the national model's tuned hyperparameters. Re-tune on the state-level panel using the same expanding-window time-series cross-validation procedure. The larger sample size (~6,000 observations) means the model can accommodate slightly more complexity — `max_depth` may be increased to 4–5 and `min_child_weight` reduced to 3–5 if validation performance improves. All other hyperparameter ranges and the tuning procedure remain as in the national model.

**Interpretation outputs:**
- SHAP main effect for $L_t$ at state level
- SHAP interaction values between $L_t$ and $D_{ki}$: direction and magnitude of the industry group modifier on the lock-in effect
- SHAP interaction values between $L_t$ and $U_{it}$: whether the unemployment interaction persists at the state level
- Partial dependence plots for $L_t$ across industry groups
- Permutation importance ranking confirming $L_t$, $U_{it}$, and $D_{ki}$ carry meaningful signal

## 6) Hyperparameter Selection

### National SARIMAX

Order selection is two-step. First, select the ARIMA order $(p, d, q)(P, D, Q)_{12}$ on $Y_t$ alone using AIC and BIC, without exogenous regressors. Second, add the full exogenous regressor set $[\widetilde{\Delta L}_t, \tilde{U}_t, \Delta H_t, \Delta S_t, \widetilde{\Delta L}_t \times \tilde{U}_t]$ and re-estimate. Retain the interaction term regardless of marginal significance. Verify residual white noise via Ljung-Box at lags 6, 12, 18, and 24. The selected order is carried forward to the state-level model unchanged.

### State SARIMAX

No order selection. The common ARIMA order from the national model is imposed directly. All regressors — state fixed effect dummies, industry group dummies, and all interaction terms — are retained as specified. If the Ljung-Box test fails under the common order, report an unconstrained version as a robustness check.

### National XGBoost

The structural hyperparameters `max_depth`, `min_child_weight`, `subsample`, and `colsample_bytree` are tuned jointly via time-series cross-validation with an expanding window (3 folds). `n_estimators` is fixed at 200 during cross-validation for stability, then early stopping is applied in the final fit on the full training set. `reg_alpha` and `reg_lambda` are fixed at 0.5 and 2.0 respectively — a second-pass regularisation search is not conducted given the ~90-row training set. All folds respect temporal order.

| Hyperparameter | Search range | Rationale |
|---|---|---|
| `max_depth` | 3, 4 | Shallow trees limit overfitting at small N |
| `min_child_weight` | 5, 10 | Prevents splits on very small node groups |
| `subsample` | 0.7, 0.8 | Row subsampling reduces variance |
| `colsample_bytree` | 0.6, 0.8 | Column subsampling reduces feature correlation |
| `learning_rate` | 0.03 (fixed) | Small rate; stable with early stopping |
| `reg_alpha` | 0.5 (fixed) | L1 regularisation |
| `reg_lambda` | 2.0 (fixed) | L2 regularisation |

### State XGBoost

Start from the national model's best hyperparameters. Re-tune `max_depth`, `min_child_weight`, `subsample`, and `colsample_bytree` on the state-level panel using the same expanding-window procedure. The larger sample (~6,000 state-month rows) means the search range can be widened slightly — `max_depth` to 4–5 and `min_child_weight` to 3–10 — if validation RMSE improves. All other settings carry over from the national model.

## 7) Evaluation Metrics

### Tier 1 — Metrics Consistent Across Both Models at Both Levels

**Root Mean Squared Error (RMSE):**

$$\text{RMSE} = \sqrt{\frac{1}{n}\sum_{t=1}^{n}(Y_t - \hat{Y}_t)^2}$$

**Mean Absolute Error (MAE):**

$$\text{MAE} = \frac{1}{n}\sum_{t=1}^{n} |Y_t - \hat{Y}_t|$$

Both computed on a **rolling expanding-window** validation scheme. Minimum training window: 60% of the full sample. At each step the window expands by one period and one out-of-sample prediction is produced. All cross-validation folds respect temporal order.

**Directional consistency check (qualitative):** across both models at both levels, verify that the sign and direction of the lock-in effect on new listing count is consistent. Agreement across all four model-level combinations constitutes strong convergent evidence.

**Relative importance ranking (qualitative):** across both models, rank $L_t$, $U_t$, $H_t$, $S_t$ by their contribution to $Y_t$. Disagreement in the ranking is itself a finding worth reporting.

### Tier 2 — Model-Specific Metrics

**SARIMAX — national:**
- $\hat{\beta}_1$ and $\hat{\beta}_5$ with 95% confidence intervals
- Ljung-Box test p-value on residuals
- AIC and BIC for model selection

**SARIMAX — state:**
- $\hat{\beta}_1$ with 95% confidence interval — average lock-in effect at the state level
- $\hat{\beta}_5$ with 95% confidence interval — unemployment interaction at the state level
- $\hat{\lambda}_k$ for each industry group with 95% confidence intervals — industry interaction effects
- Ljung-Box test p-value on residuals

**XGBoost — national:**
- SHAP main effect for $L_t$: mean absolute SHAP value across test set
- SHAP interaction values between $L_t$ and $U_t$
- Partial dependence plots for $L_t$ and $L_t \times U_t$
- Permutation importance ranking

**XGBoost — state:**
- SHAP main effect for $L_t$ at state level
- SHAP interaction values between $L_t$ and $D_{ki}$
- SHAP interaction values between $L_t$ and $U_{it}$
- Partial dependence plots for $L_t$ across industry groups
- Permutation importance ranking
